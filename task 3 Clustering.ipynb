```
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```
#Getting the Iris_Dataset Features
from sklearn import datasets
iris= datasets.load_iris()
iris_df=pd.DataFrame(data=(iris.data),columns=(iris.feature_names))
```

```
iris_df.head()
```

```
df=iris_df.iloc[:,[0,1,2,3]].values
```

```
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(df,method="ward"))
plt.figure(1, figsize=(10, 100)) # Figure size in inches (size_x, size_y)
#plt.subplot(211)
plt.title("Dendrogram")
plt.xlabel("Iris")
plt.ylabel("Euclidean disances")
plt.show()
```

```
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 3, affinity = 'euclidean', linkage = 'ward')
y_hc = hc.fit_predict(df)
```

```
# Visualising the clusters
plt.scatter(df[y_hc == 0, 0], df[y_hc == 0, 1], s = 100, c = 'red', label = 'Iris-setosa')
plt.scatter(df[y_hc == 1, 0], df[y_hc == 1, 1], s = 100, c = 'blue', label = 'Iris-versicolour')
plt.scatter(df[y_hc == 2, 0], df[y_hc == 2, 1], s = 100, c = 'green', label = 'Iris-virginica')
plt.legend()
plt.show()
```
